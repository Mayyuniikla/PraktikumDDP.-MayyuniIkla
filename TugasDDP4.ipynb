```python
# Meminta pengguna memasukkan sebuah bilangan bulat
bilangan = int(input("Masukkan sebuah bilangan bulat: "))

# Menentukan apakah bilangan tersebut genap atau ganjil
if bilangan % 2 == 0:
    print(f"{bilangan} adalah bilangan genap.")
else:
    print(f"{bilangan} adalah bilangan ganjil. ")
```

```python

# Masukkan nilai ujian
nilai = float(input("Masukkan nilai anda: "))

# Menentukan apakah nilai lulus atau tidak
if nilai >= 75:
    print("Lulus")
else:
    print("Tidak Lulus")
```

```python
#Masukkan usia
usia = int(input("Masukkan usia: "))

# Menentukan kategori berdasarkan usia
if usia >= 18:
    print("Kamu Dewasa")
elif usia >= 13 and usia <= 17:
    print("Remaja")
else:
    print("Anak-anak")
```

```python
#Masukkan status keanggotaan
status = input("Masukkan status keanggotaan (gold, silver, atau bronze): ").lower()

# Memeriksa status keanggotaan dan memberikan respon sesuai
if status == "gold" or status == "silver":
    print("Selamat! Anda mendapatkan diskon.")
else:
    print("Maaf, Anda tidak mendapatkan diskon.")
```

```python

# memasukkan total pembelian
total_pembelian = float(input("Masukkan total pembelian: "))

# Menentukan diskon jika pembelian di atas 100
if total_pembelian > 100:
  diskon = total_pembelian * 0.1  # Diskon 10%
  total_bayar = total_pembelian - diskon
  print("Selamat anda mendapatkan diskon!")
  print("Total yang harus dibayar:", total_bayar)
else:
  print("Total yang harus dibayar:", total_pembelian)
```
